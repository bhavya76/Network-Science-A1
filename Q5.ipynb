```python
import networkx as nx
import matplotlib.pyplot as plt

with open('road-euroroad.edges', 'r') as f:
    next(f)
    next(f)
    edges = [tuple(map(int, line.strip().split()[:2])) for line in f]

G = nx.Graph(edges)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=50, font_size=1)
plt.savefig('euro-road-spring.png')
plt.show()
print()
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_size=50, font_size=1)
plt.savefig('euro-road-circular.png')
plt.show()
print()
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, with_labels=True, node_size=50, font_size=1)
plt.savefig('euro-road-kamada.png')
plt.show()
```
